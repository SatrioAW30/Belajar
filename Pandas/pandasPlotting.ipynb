Plotting With Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('PriceHouse.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.plot()
```

```python
df['price'].plot()
```

```python
df['price'].plot(legend=True)
```

```python
df.plot(x='price', y='bedrooms', kind='scatter', legend=True)
```

```python
df.plot(title='Sample Plot',figsize=(15,10))
```

```python
df['price'].plot(kind='box',title='Box Plot')
```

```python
df['price'].plot(kind='hist',title='Histogram')
```
